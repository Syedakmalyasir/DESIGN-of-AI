```python
out1=np.where(data["SepalWidthCm"]>4)
out1
```

```python
m=data["SepalWidthCm"].mean(axis=0)
```

```python
l=[]
for i in out1:
  l.append(data["SepalWidthCm"][i])
```

```python
data["SepalWidthCm"]=data["SepalWidthCm"].replace(to_replace=l,value=m)
```
